```python
def maxSubArray(nums):
    global_max = local_max = nums[0]
    for n in nums[1:]:
        local_max  = max(local_max + n, n)
        global_max = max(local_max, global_max)
    return global_max 
```

```python
cases = [[-2,1,-3,4,-1,2,1,-5,4], [1], [0], [-1], [-2147483647, 1], [2147483647, 1]]
[maxSubArray(L) for L in cases]
```
